```python
import pandas as pd
import os

# Create folders using relative paths
os.makedirs('../raw_data', exist_ok=True)
os.makedirs('../processed_data', exist_ok=True)

# Create a messy dictionary (The "Dirty" Data)
data = {
    'Date': ['2026-05-01', '05/02/2026', 'May 3, 2026', '2026-05-04'],
    'Employee': [' Wilton', 'GARY', 'wilton', 'Maam Lou'],
    'Sales': ['1500', '2100', 'unknown', -500],
    'Dept': ['IT', 'it', 'IT', 'Admin']
}

# Save it as an Excel file
df_raw = pd.DataFrame(data)
df_raw.to_excel('../raw_data/staff_logs.xlsx', index=False)

print("Step 1 Complete: Messy file created in /raw_data/")
```

```python
# 1. Load the data
df = pd.read_excel('../raw_data/staff_logs.xlsx')

# 2. Logic: Normalize Names (Fix spaces and casing)
df['Employee'] = df['Employee'].str.strip().str.title()

# 3. Logic: Fix Dates (Standardize different formats)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# 4. Logic: Numeric Cleaning
# Turn 'unknown' into 0 and remove negative values (Business Rule)
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0)
df = df[df['Sales'] >= 0]

# 5. Display the result
df
```

```python
df.to_excel('../processed_data/final_report.xlsx', index=False)
print("Step 3 Complete: Clean report saved in /processed_data/")
```
