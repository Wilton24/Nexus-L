```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
response = requests.get(url)

if response.status_code == 200:
    # Parse the raw HTML text into a searchable tree structure
    soup = BeautifulSoup(response.text, 'html.parser')
    print("SUCCESS: HTML downloaded and parsed into the soup engine.")
else:
    print(f"FAILED to fetch. Status code: {response.status_code}")
```

```python
# 1. Find the very first book card on the page
first_book = soup.find('article', class_='product_pod')

if first_book:
    # 2. Extract the raw title from the anchor <a> tag inside the <h3> tag
    raw_title = first_book.h3.a['title']
    
    # 3. Extract the raw price string from the paragraph element
    raw_price = first_book.find('p', class_='price_color').text
    
    print("--- RAW EXTRACTION ---")
    print(f"Book:  {raw_title}")
    print(f"Price: {raw_price}")
    
    # --- DATA ENGINEERING / TRANSFORMATION ---
    # The price comes as "£51.77". We need to strip the currency symbol '£' 
    # so we can convert it to a floating-point number.
    clean_price = raw_price.replace('£', '')
    numeric_price = float(clean_price)
    
    print("\n--- CLEANED DATA ---")
    print(f"Price Type: {type(numeric_price)}")
    print(f"Price Value: {numeric_price}")
else:
    print("Could not find any books matching that selector.")
```
